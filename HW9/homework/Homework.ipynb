### Step 0

Run the necessary imports.

```python
import cv2
import dlib
from matplotlib import pyplot as plt
```

### Step 1

Load an image (any image that contains faces).

```python
# Load image containing faces
img = cv2.imread('../data/eu_summit.jpg')

# Convert BGR to RGB for matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Input image")
plt.axis("off")
```

### Step 2

Load the dlib face predictor.

```python
# Load the frontal face detector from dlib
detector = dlib.get_frontal_face_detector()
```

### Step 3

Run the predictor on your image.

```python
# Convert image to grayscale for detection
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces
faces = detector(gray, 1)

print("Number of faces detected:", len(faces))
```

### Step 4

Draw bounding boxes around the detected faces and plot the image. Use different colour for each face.

```python
import random

# Copy image for visualization
vis = img_rgb.copy()

# Draw rectangle for each detected face
for i, rect in enumerate(faces):

    x = rect.left()
    y = rect.top()
    w = rect.right() - x
    h = rect.bottom() - y

    # Generate random color
    color = (
        random.randint(0,255),
        random.randint(0,255),
        random.randint(0,255)
    )

    cv2.rectangle(vis, (x,y), (x+w,y+h), color, 3)

# Show result
print("Faces drawn:", len(faces))

plt.figure(figsize=(10,6))
plt.imshow(vis)
plt.title("Detected faces")
plt.axis("off")
```

### Step 5 (optional)

Repeat the process with a different and more challenging image (more faces, smaller faces, people with glasses, hats, helmets, etc.). How does the detector perform? Is it robust?

```python
img2 = cv2.imread('../data/zelensky.jpg')
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

faces2 = detector(gray2,1)

vis2 = img2_rgb.copy()

for rect in faces2:

    x = rect.left()
    y = rect.top()
    w = rect.right() - x
    h = rect.bottom() - y

    cv2.rectangle(vis2, (x,y), (x+w,y+h), (255,0,0), 3)

plt.imshow(vis2)
plt.title("Optional test image")
plt.axis("off")
```
